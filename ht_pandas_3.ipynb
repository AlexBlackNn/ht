## Расчет

```python
import pandas as pd
```

```python
log_df = pd.read_csv('/content/visit_log.csv',sep=';')
log_df.tail()
```

```python
def create_source_type(df):
  if df['traffic_source'] == 'yandex' or df['traffic_source'] == 'google':
    return 'organic'
  elif df['region'] == 'Russia' and (df['traffic_source'] == 'paid' or df['traffic_source'] == 'email'):
    return 'ad'
  elif df['traffic_source'] == 'paid' or df['traffic_source'] == 'email':
    return 'other'
  else:
    return df['traffic_source']          
```

```python
log_df['source_type'] = log_df.apply(create_source_type, axis = 1)
log_df.tail(20)
```

# Задание 3

  Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.


```python
rating_df = pd.read_csv('/content/ml-latest-small/ratings.csv')
rating_df['timestamp'] =  pd.to_datetime(rating_df['timestamp'],unit="s") 
rating_df
```

```python
aggregate_df = rating_df.groupby(by = 'userId').agg({'timestamp': ['min', 'max'], 'movieId': 'count'})
aggregate_df = aggregate_df[aggregate_df['movieId']['count'] > 100]
aggregate_df['lifetime'] = aggregate_df['timestamp']['max'] - aggregate_df['timestamp']['min']
aggregate_df
```

```python
print('Cреднее время жизни', aggregate_df['lifetime'].mean())
```

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

## Первая таблица

```python
rzd_auto_df = rzd.merge(auto, how='outer')
rzd_auto_air_df = rzd_auto_df.merge(air, how='outer')
rzd_auto_air_df
```

## Вторая таблица

```python
rzd_auto_air_address_df = rzd_auto_air_df.merge(client_base, how='outer')
rzd_auto_air_address_df
```
